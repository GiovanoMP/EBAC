#### 1. Carregue os dados da tarefa anterior

````python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
````

````python
file_path = "dados_ativos2.csv"
df = pd.read_csv(file_path, skiprows=1)  # Pular a primeira linha desnecessária

# Ajustar as colunas com base na segunda linha
colunas = ['Date', 'AAPL_Adj_Close', 'GOOGL_Adj_Close', 'MSFT_Adj_Close', 
           'AAPL_Close', 'GOOGL_Close', 'MSFT_Close', 
           'AAPL_High', 'GOOGL_High', 'MSFT_High', 
           'AAPL_Low', 'GOOGL_Low', 'MSFT_Low', 
           'AAPL_Open', 'GOOGL_Open', 'MSFT_Open', 
           'AAPL_Volume', 'GOOGL_Volume', 'MSFT_Volume']
df.columns = colunas

# Remover as linhas que não contêm dados úteis
df = df.drop([0]).reset_index(drop=True)

# Converter a coluna 'Date' para datetime
df['Date'] = pd.to_datetime(df['Date'])

# Definir 'Date' como índice
df.set_index('Date', inplace=True)
````

2.1 Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

````python
# Calcular a média móvel de 90 dias para os preços de fechamento ajustados
janela = 90
df['AAPL_Media'] = df['AAPL_Adj_Close'].astype(float).rolling(window=janela).mean()
df['GOOGL_Media'] = df['GOOGL_Adj_Close'].astype(float).rolling(window=janela).mean()
df['MSFT_Media'] = df['MSFT_Adj_Close'].astype(float).rolling(window=janela).mean()
````

2.2 Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
# Calcular o desvio padrão de 90 dias para os preços de fechamento ajustados
df['AAPL_Desvio'] = df['AAPL_Adj_Close'].astype(float).rolling(window=janela).std()
df['GOOGL_Desvio'] = df['GOOGL_Adj_Close'].astype(float).rolling(window=janela).std()
df['MSFT_Desvio'] = df['MSFT_Adj_Close'].astype(float).rolling(window=janela).std()
````

2.3 Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
# Plotar as médias móveis com Plotly
fig_media = go.Figure()
fig_media.add_trace(go.Scatter(x=df.index, y=df['AAPL_Media'], mode='lines', name='AAPL Média Móvel'))
fig_media.add_trace(go.Scatter(x=df.index, y=df['GOOGL_Media'], mode='lines', name='GOOGL Média Móvel'))
fig_media.add_trace(go.Scatter(x=df.index, y=df['MSFT_Media'], mode='lines', name='MSFT Média Móvel'))

fig_media.update_layout(
    title='Média Móvel de 90 dias dos Preços de Fechamento Ajustados',
    xaxis_title='Data',
    yaxis_title='Preço de Fechamento Ajustado'
)
fig_media.show()
````

4. Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
# Plotar os desvios padrão com Plotly
fig_desvio = go.Figure()
fig_desvio.add_trace(go.Scatter(x=df.index, y=df['AAPL_Desvio'], mode='lines', name='AAPL Desvio Padrão'))
fig_desvio.add_trace(go.Scatter(x=df.index, y=df['GOOGL_Desvio'], mode='lines', name='GOOGL Desvio Padrão'))
fig_desvio.add_trace(go.Scatter(x=df.index, y=df['MSFT_Desvio'], mode='lines', name='MSFT Desvio Padrão'))

fig_desvio.update_layout(
    title='Desvio Padrão de 90 dias dos Preços de Fechamento Ajustados',
    xaxis_title='Data',
    yaxis_title='Desvio Padrão'
)
fig_desvio.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Inspecionar as colunas do DataFrame
print(df.columns)
````

**Obs:** Tive vários problemas com a coluna 'Date', deta forma, utilizei os recursos abaixo para conseguir plotar o gráfico de candlestick.

````python
# Converter a coluna 'Date' para datetime, se não estiver presente
if 'Date' not in df.columns:
    df.insert(0, 'Date', pd.date_range(start='2022-01-03', periods=len(df), freq='B'))

# Ajustar as colunas do DataFrame conforme necessário
df.columns = ['Date', 'AAPL_Adj_Close', 'GOOGL_Adj_Close', 'MSFT_Adj_Close', 
              'AAPL_Close', 'GOOGL_Close', 'MSFT_Close', 'AAPL_High', 'GOOGL_High', 
              'MSFT_High', 'AAPL_Low', 'GOOGL_Low', 'MSFT_Low', 'AAPL_Open', 
              'GOOGL_Open', 'MSFT_Open', 'AAPL_Volume', 'GOOGL_Volume', 'MSFT_Volume', 
              'AAPL_Media', 'GOOGL_Media', 'MSFT_Media', 'AAPL_Desvio', 'GOOGL_Desvio', 
              'MSFT_Desvio']

# Converter a coluna 'Date' para datetime
df['Date'] = pd.to_datetime(df['Date'])

# Selecionar os últimos 60 dias corridos
df_60d = df[df['Date'] >= (df['Date'].max() - pd.Timedelta(days=60))]

# Criar o gráfico de candlestick para o ativo "AAPL"
fig = go.Figure(data=[go.Candlestick(x=df_60d['Date'],
                                     open=df_60d['AAPL_Open'],
                                     high=df_60d['AAPL_High'],
                                     low=df_60d['AAPL_Low'],
                                     close=df_60d['AAPL_Close'])])

# Adicionar título e ajustar layout
fig.update_layout(title='Gráfico de Candlestick - AAPL (Últimos 60 dias)',
                  xaxis_title='Data',
                  yaxis_title='Preço',
                  xaxis_rangeslider_visible=False)

# Exibir o gráfico
fig.show()
````
