```python
# Carregar os dados do arquivo CSV
dados = pd.read_csv('dados_ativos.csv', header=[0, 1], index_col=0, parse_dates=True)

# Verificar as primeiras linhas do DataFrame
print(dados.head())
```

### 2.1 Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).

```python
# Calcular a amplitude de variação dentro do dia
amplitude_variacao = dados['High'] - dados['Low']

# Exibindo as primeiras linhas para verificação
print(amplitude_variacao.head())
```

### 2.2 Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.

```python
# Criando o gráfico de linha usando plotly.graph_objects
fig = go.Figure()

# Adicionando traces para cada ativo
for ativo in ativos:
    fig.add_trace(go.Scatter(x=amplitude_variacao.index, y=amplitude_variacao[ativo], mode='lines', name=ativo))

# Configurando o layout do gráfico
fig.update_layout(
    title='Variação da Amplitude ao Longo do Período',
    xaxis_title='Data',
    yaxis_title='Amplitude',
    legend_title='Ativo'
)

# Exibindo o gráfico
fig.show()
```

### 2.3 Veja se consegue identificar que ativo apresentou maior volatilidade.

Embora possamos verificar visualmente e concluir que o ativo da **Microsoft** é o que possui maior volatilidade, como exemplificação, abaixo, também podemos verificar programaticamente em cenários mais complexos.
 

```python
# Calcular a volatilidade (desvio padrão da amplitude) de cada ativo
volatilidade_aapl = amplitude_variacao['AAPL'].std()
volatilidade_msft = amplitude_variacao['MSFT'].std()
volatilidade_googl = amplitude_variacao['GOOGL'].std()

# Identificar o ativo com maior volatilidade
volatilidades = {'AAPL': volatilidade_aapl, 'MSFT': volatilidade_msft, 'GOOGL': volatilidade_googl}
ativo_maior_volatilidade = max(volatilidades, key=volatilidades.get)

# Exibir as volatilidades e o ativo com maior volatilidade
print(f'Volatilidade AAPL: {volatilidade_aapl}')
print(f'Volatilidade MSFT: {volatilidade_msft}')
print(f'Volatilidade GOOGL: {volatilidade_googl}')
print(f'Ativo com maior volatilidade: {ativo_maior_volatilidade}')
```
