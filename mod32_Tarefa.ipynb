1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
import numpy as np
from scipy import stats

# Parmetros da distribuio normal
media = 15  # mdia dos pacotes em kg
desvio_padrao = 0.1  # desvio padro em kg

# Peso alvo que queremos calcular a probabilidade
peso_alvo = 15.2

# Clculo da probabilidade acumulada at 15.2 kg
# A funo norm.cdf calcula a probabilidade acumulada at um valor especfico
# loc  a mdia da distribuio e scale  o desvio padro
probabilidade_acumulada = stats.norm.cdf(peso_alvo, loc=media, scale=desvio_padrao)

# Para obter a probabilidade de ser maior que 15.2 kg, subtramos a probabilidade acumulada de 1
probabilidade = 1 - probabilidade_acumulada

# Exibindo o resultado
print(f"A probabilidade de que um pacote tenha mais de {peso_alvo} kg  de {probabilidade:.4f} ou {probabilidade*100:.2f}%.")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parmetros da distribuio normal
media = 15  # mdia dos pacotes em kg
desvio_padrao = 0.1  # desvio padro em kg

# Percentil desejado
percentil = 0.95 

# Clculo do valor correspondente ao percentil de 95%
# A funo norm.ppf calcula o valor da varivel aleatria que corresponde  probabilidade acumulada fornecida
valor_percentil_95 = stats.norm.ppf(percentil, loc=media, scale=desvio_padrao)

# Exibindo o resultado
print(f"O peso correspondente ao percentil de 95%  de {valor_percentil_95:.4f} kg.")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Gerando valores de Y
Y = np.linspace(-6, 6, 200)

# Clculo das probabilidades para Y_1 (Distribuio Normal Padro)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)

# Clculo das probabilidades para Y_2 (Distribuio t-Student com 5 graus de liberdade)
prob_Y2 = stats.t.cdf(Y, df=5)

# Exibindo os resultados
print("Probabilidades para Y_1 (Normal Padro):")
print(prob_Y1)

print("\nProbabilidades para Y_2 (t-Student com 5 graus de liberdade):")
print(prob_Y2)
```

### Probabilidades Acumuladas para \( Y_1 \) e \( Y_2 \)

- **\( Y_1 \)**: Distribuio **Normal Padro** (mdia = 0, desvio padro = 1).
- **\( Y_2 \)**: Distribuio **t-Student** com 5 graus de liberdade.

#### Resultados

1. **\( Y_1 \) (Normal Padro)**:
   - Para \( Y \) muito negativos (\(< -4\)), a probabilidade acumulada  prxima de 0.
   - Para \( Y \) muito positivos (\(> 4\)), a probabilidade acumulada  prxima de 1.

2. **\( Y_2 \) (t-Student com 5 graus de liberdade)**:
   - Apresenta caudas mais largas que a normal, resultando em probabilidades acumuladas menores nos extremos.
   - Para \( Y \) perto de 0, as probabilidades acumuladas de \( Y_1 \) e \( Y_2 \) so semelhantes.



4. Faa uma visualizao grfica dos dados do item anterior.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando valores de Y
Y = np.linspace(-6, 6, 200)

# Clculo das probabilidades acumuladas
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)
prob_Y2 = stats.t.cdf(Y, df=5)

# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_Y1, label='Normal Padro (Y1)', color='blue', linestyle='-')
plt.plot(Y, prob_Y2, label='t-Student com 5 GL (Y2)', color='red', linestyle='--')

# Adicionando ttulo e rtulos
plt.title('Comparao das Probabilidades Acumuladas (CDF) para Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')
plt.legend()
plt.grid(True)

# Exibindo o grfico
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Gerando valores de P
P = np.linspace(0, 1, 20)

# Clculo dos percentis para Y_1 (Distribuio Normal Padro)
percentis_Y1 = stats.norm.ppf(P, loc=0, scale=1)

# Clculo dos percentis para Y_2 (Distribuio t-Student com 5 graus de liberdade)
percentis_Y2 = stats.t.ppf(P, df=5)

# Exibindo os resultados
print("Percentis para Y_1 (Normal Padro):")
print(percentis_Y1)

print("\nPercentis para Y_2 (t-Student com 5 graus de liberdade):")
print(percentis_Y2)
```

### Resumo dos Resultados

1. **Probabilidade de um Pacote ter mais de 15,2 kg**:
   - A probabilidade  de **0,0228** ou **2,28%**.

2. **Peso Correspondente ao Percentil de 95%**:
   - O peso que corresponde ao percentil de 95%  **15,1645 kg**.

3. **Probabilidades Acumuladas para \( Y_1 \) (Normal Padro)**:
   - As probabilidades acumuladas (CDF) variam de aproximadamente **\(9.87 \times 10^{-10}\)** para \( Y = -6 \) at quase **1** para \( Y = 6 \).
   - Essas probabilidades seguem uma distribuio simtrica, centrada em 0, refletindo a simetria da distribuio normal.

4. **Probabilidades Acumuladas para \( Y_2 \) (t-Student com 5 graus de liberdade)**:
   - As probabilidades acumuladas (CDF) variam de aproximadamente **\(9.23 \times 10^{-4}\)** para \( Y = -6 \) at quase **1** para \( Y = 6 \).
   - A distribuio *t-Student* tem caudas mais pesadas, resultando em probabilidades acumuladas mais baixas nos extremos em comparao com a distribuio normal.

5. **Percentis para \( Y_1 \) (Normal Padro)**:
   - Os percentis calculados para \( Y_1 \) refletem os valores da distribuio normal padro para as probabilidades de 0 a 1, variando de **-\(\infty\)** a **\(\infty\)** com percentis intermedirios como \( -1.62 \), \( 0 \), e \( 1.62 \).

6. **Percentis para \( Y_2 \) (t-Student com 5 graus de liberdade)**:
   - Os percentis para \( Y_2 \) tambm variam de **-\(\infty\)** a **\(\infty\)**, mas os valores so mais extremos que na distribuio normal, com percentis intermedirios como \( -1.97 \), \( 0 \), e \( 1.97 \), refletindo as caudas mais pesadas da distribuio *t-Student*.

Esses resultados ilustram como as distribuies Normal Padro e *t-Student* com 5 graus de liberdade diferem em termos de probabilidade acumulada e percentis, especialmente nos extremos.


6. Faa um grfico com os dados do item anterior.

```python
# Gerando valores de P
P = np.linspace(0, 1, 20)

# Clculo dos percentis para Y_1 (Distribuio Normal Padro)
percentis_Y1 = stats.norm.ppf(P, loc=0, scale=1)

# Clculo dos percentis para Y_2 (Distribuio t-Student com 5 graus de liberdade)
percentis_Y2 = stats.t.ppf(P, df=5)

# Criando o grfico
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_Y1, label='Normal Padro (Y1)', color='blue', linestyle='-')
plt.plot(P, percentis_Y2, label='t-Student 5 GL (Y2)', color='red', linestyle='--')

# Adicionando ttulo e rtulos
plt.title('Comparao dos Percentis para Y1 e Y2')
plt.xlabel('Probabilidade (P)')
plt.ylabel('Percentil')
plt.legend()
plt.grid(True)

# Exibindo o grfico
plt.show()
```
